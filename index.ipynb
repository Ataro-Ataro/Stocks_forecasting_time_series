```python
df = df.dropna()
df.info()
```

```python
# rename stock code colum to Code fro easier join
df = df.rename(columns={'Stock_code' : 'Code'})

df.head()
```

```python
df.columns
```

```python
# add sectors column to merged df
df = df[['Code', 'Sector']]

merged_df = pd.merge(merged_df, df, on='Code', how='left')

merged_df
```

```python
merged_df.isnull().sum()
```

```python
merged_df[merged_df.isnull().any(axis=1)] 
```

```python
merged_df.loc[merged_df['Code'] == '^NBDI', 'Sector'] = 'Indices'

# Fill 'Sector' for 'HFCK-R' as 'Banking'
merged_df.loc[merged_df['Code'] == 'HFCK-R', 'Sector'] = 'Banking'

merged_df[merged_df.isnull().any(axis=1)] 
```

```python
# replace nan values with 'Indices'
merged_df.fillna({'Sector' : 'Indices'}, inplace=True)

merged_df.info()
```

```python
# convert hyphens in volume column to 0
merged_df['Volume'] = merged_df['Volume'].astype(str).str.replace(',', '', regex=False).replace('-', np.nan)
merged_df['Volume'] = pd.to_numeric(merged_df['Volume'], errors='coerce')
merged_df.fillna({'Volume' : 0}, inplace=True)
merged_df
```

```python
# convert hyphens in change to 0
merged_df['Change'] = merged_df['Volume'].astype(str).str.replace(',', '', regex=False).replace('-', np.nan)
merged_df['Change'] = pd.to_numeric(merged_df['Change'], errors='coerce')
merged_df.fillna({'Change' : 0}, inplace=True)
merged_df
```

```python
# convert hyphens in change% into 0%
merged_df['Change%'] = merged_df['Change%'].replace('-', '0.0%')
merged_df['Change%'] =  merged_df['Change%'].astype(str).str.replace('%', '')
merged_df['Change%'] = pd.to_numeric(merged_df['Change%'])
merged_df.info()
```

```python
# drop adjusted price column
merged_df.drop('Adjusted Price', axis=1, inplace=True)
merged_df.head()
```

```python
# Convert 'Date' to datetime
merged_df['Date'] = pd.to_datetime(merged_df['Date'], format='mixed')
merged_df.head()
```

```python
#convert price columns to numeric
price_columns = ['12m Low', '12m High', 'Day Low', 'Day High', 'Day Price', 'Previous']

for column in price_columns:
    merged_df[column] = merged_df[column].astype(str).str.replace(',', '', regex=False)
    merged_df[column] = pd.to_numeric(merged_df[column], errors='coerce')
    merged_df.fillna({column : 0}, inplace=True)

merged_df.info()
```

### EXPLORATORY DATA ANALYSIS

```python
# sectors distribution
plt.figure(figsize=(12, 8))
sector_counts = merged_df['Sector'].value_counts()
sns.barplot(x=sector_counts.values, y=sector_counts.index, hue=sector_counts.index, palette='viridis')
plt.title('Number of Stocks per Sector')
plt.xlabel('Number of Stocks')
plt.ylabel('Sector')
plt.tight_layout()
```

```python
# Sector-wise Average Price
plt.figure(figsize=(12, 8))
sector_avg_price = merged_df.groupby('Sector')['Day Price'].mean().sort_values(ascending=False)
sns.barplot(x=sector_avg_price.values, y=sector_avg_price.index, hue=sector_avg_price.index, palette='plasma')
plt.title('Average Stock Price per Sector')
plt.xlabel('Average Day Price')
plt.ylabel('Sector')
plt.tight_layout()
```

* Although commercial and services has the highest number of stocks, it is among the lowest average stock prices. This could indicate a sector with many smaller companies or companies with lower individual stock valuations.
* Telecommunication sector has a moderate number of stocks but boasts one of the highest average stock prices suggesting that the companies in this sector are generally more highly valued.

```python
# Stock Price Evolution for Sample Stocks
sample_stocks = {
    'SCOM': 'Telecommunication and Technology',
    'EQTY': 'Banking',
    'BAMB': 'Construction and Allied'
}

plt.figure(figsize=(15, 8))
for code, sector in sample_stocks.items():
    stock_data = merged_df[merged_df['Code'] == code]
    plt.plot(stock_data['Date'], stock_data['Day Price'], label=f'{code} ({sector})')

plt.title('Stock Price Evolution for Sample Stocks')
plt.xlabel('Date')
plt.ylabel('Day Price')
plt.legend()
plt.grid(True)
plt.tight_layout()
```

```python
# top 10 most volatile stocks
stock_volatility = merged_df.groupby('Name')['Change%'].std().sort_values(ascending=False) * (252**0.5)

# Visualize top 10 most volatile stocks
plt.figure(figsize=(12, 7))
sns.barplot(x=stock_volatility.head(10).values, y=stock_volatility.head(10).index, hue=stock_volatility.head(10), palette='viridis')
plt.title('Top 10 Most Volatile Stocks (Annualized Volatility of Daily Returns)')
plt.xlabel('Annualized Volatility (%)')
plt.ylabel('Stock Name')
plt.tight_layout()
plt.show()
```

```python
# function to print codes with their sectors
def sectors_for_codes(codes_list, dataframe, code_col='Code', sector_col='Sector'):
    # Create a mapping from code to sector
    code_sector_mapping = dataframe.set_index(code_col)[sector_col].to_dict()

    print(f'Sectors for the Provided Stock Codes:')
    print(f'---------------------------------------------')
    for code in codes_list:
        sector = code_sector_mapping.get(code)
        print(f'{code}: {sector}')
```

```python
# List of the top 10 volatile stock codes
volatile_codes = ['^N10I', '^N20I', 'HFCK-R', 'PORT', 'IMH', 'SLAM', 'TPSE', 'OCH', 'SMER', 'NBV']

sectors_for_codes(volatile_codes, merged_df, code_col='Code', sector_col='Sector')
```

* Indicies and Banking sectors are prominent among the sectors with highest volatility

```python
first_prices = merged_df.groupby('Code')['Day Price'].first()
last_prices = merged_df.groupby('Code')['Day Price'].last()

# Calculate total return
total_returns = ((last_prices - first_prices) / first_prices) * 100
total_returns = total_returns.replace([float('inf'), -float('inf')], pd.NA).dropna()
top_returns = total_returns.sort_values(ascending=False)

# Visualize top 10 performers
plt.figure(figsize=(12, 7))
sns.barplot(x=top_returns.head(10).values, y=top_returns.head(10).index, hue=top_returns.head(10).index, palette='viridis')
plt.title('Top 10 Stocks by Total Return (%)')
plt.xlabel('Total Return (%)')
plt.ylabel('Stock Code')
plt.tight_layout()
plt.show()
```

```python
# List of the top 10 stocks with their sector by total returN
codes = ['ORCH', 'PORT', 'KPLC', 'UMME', 'IMH', 'KAPC', 'SCBK', 'CARB', 'BAMB', 'EVRD']

sectors_for_codes(codes, merged_df, code_col='Code', sector_col='Sector')
```

* While a few stocks delivered extraordinary returns(ORCH stocks), several sectors particularly Construction and Allied, showed consistent strong performance among the top gainers.

```python
# Visualize bottom 10 performers
plt.figure(figsize=(12, 7))
sns.barplot(x=top_returns.tail(10).values, y=top_returns.tail(10).index, hue=top_returns.tail(10).index, palette='viridis')
plt.title('Bottom 10 Stocks by Total Return (%)')
plt.xlabel('Total Return (%)')
plt.ylabel('Stock Code')
plt.tight_layout()
plt.show()
```

```python
# top 10 stocks by daily trading volume
stock_avg_volume = merged_df.groupby('Code')['Volume'].mean().sort_values(ascending=False)

plt.figure(figsize=(12, 7))
sns.barplot(x=stock_avg_volume.head(10).values, y=stock_avg_volume.head(10).index, hue=stock_avg_volume.head(10).index, palette='viridis')
plt.title('Top 10 Stocks by Average Daily Trading Volume')
plt.xlabel('Average Volume')
plt.ylabel('Stock Code')
plt.tight_layout()
plt.show()
```

```python
# top 10 stocks by daily trading volume with their sectors
codes = ['SCOM', 'HFCK-R', 'KNRE', 'EQTY', 'KCB', 'KEGN', 'KPLC', 'COOP', 'ABSA', 'BRIT']
sectors_for_codes(codes, merged_df, code_col='Code', sector_col='Sector')
```

*  The most actively traded stocks are heavily concentrated within the financial services sector (banking and insurance) followed by the construction and allied industries with telecommunications also showing strong individual performance. This composition often reflects sectors that are central to the economy and attract consistent investor participation.
